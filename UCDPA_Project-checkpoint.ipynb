```python
# check for missing values

def fun_missing_val():
    val = house_data_df.isnull().sum()
    print(val)

fun_missing_val()
#house_data_df.isnull().sum()
```

```python
## Here we will check the percentage of nan values present in each feature
## 1 -step make the list of features which has missing values
features_na=[features for features in house_data_df.columns if house_data_df[features].isnull().sum()>1]
## 2- step print the feature name and the percentage of missing values

for feature in features_na:
    print(feature, np.round(house_data_df[feature].isnull().mean(), 4),  ' % missing values')
```

# Duplicate Values handling 

```python
# Checking and removing duplicate records from dataset
print(house_data_df[house_data_df.duplicated()])
   

# remove duplicate records from dataset
     
house_data_df=house_data_df.drop_duplicates(keep=False)
print(house_data_df)
  
```

# Since they missing values, we need to find the relationship between missing values and Sales Price

 below diagram for this relationship

```python


for feature in features_na:
    data = house_data_df.copy()
    
    # let's make a variable that indicates 1 if the observation was missing or zero otherwise
    data[feature] = np.where(data[feature].isnull(), 1, 0)
    
    # let's calculate the mean SalePrice where the information is missing or present
    data.groupby(feature)['price'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

# Spliting the data into target/test 

```python
X = house_data_df.drop(['price'], axis=1)
Y = house_data_df['price']
print(X.shape)
print(Y.shape)
```

# Spliting data into training and testing 

```python

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=2)

## checking data after split training and testing dataset 
print(X_train.shape,Y_train.shape,X_test.shape,Y_test.shape)
```

# Model Training with (XGBoost Regressor)

     

```python
# loding Model
#from xgboost import XGBRegressor

Model = XGBRegressor()
print (Model)
#Training model with X_trann data.

Model.fit(X_train,Y_train)
```

```python
xgb.plot_importance(Model,ax=plt.gca())
```

### Evaluation & prediction  on Training Dataset

```python
# Accuracy checking for prediction  on Traning date set

Training_data_set = Model.predict(X_train)
Tbl_output=pd.DataFrame(Training_data_set)
Tbl_output.columns= ['price_pridtion']
# Below records are the prediction  from x_train data 
#print(Training_data_set)
print(Tbl_output.head())
```

```python
# R-squared score 

Pridtcion_score_R2 = metrics.r2_score(Y_train,Training_data_set)

# Mean Absoulte  Error 

Pridtcion_score_mean = metrics.mean_absolute_error(Y_train,Training_data_set)

print('R Squared score :', Pridtcion_score_R2 )

print('Mean Absoulte  Error :', Pridtcion_score_mean )
```
